```python
features = np.random.randn(100,5)
target = np.random.randn(100) > 0.5

print(features.shape, features[:5])
print(target.shape, target[:5])
```

```python
N = features.shape[0]
N_train = int(np.floor(0.7 * N))

print(N)
print(N_train)
```

```python
idx = np.random.permutation(N) # Randomizes index
print(idx.shape)
print(idx)
```

```python
idx_train = idx[:N_train]
idx_test = idx[N_train:]
print(idx_train)
print(idx_test)
```

```python
# Split data into training and testing sets
features_train = features[idx_train,:]
target_train = target[idx_train]
features_test = features[idx_test,:]
target_test = target[idx_test]

print(features_train.shape)
print(target_train.shape)
print(features_test.shape)
print(target_test.shape)
```

使用 Cross-Validation(CV) 要注意一些事項：

1. CV 假設了一個前提：訓練集資料分布「代表」了測試集資料分佈。如果否，CV 得到的誤差與預測結果可能會太樂觀了。因此，要確保盡量減小訓練集與測試集之間，資料分布的差異性。
2. 某些資料可能使用了暫時性的特徵：譬如，要使用上個月收入來預測這個月收入，那麼我們要確保在建立訓練集資料或是每次做 k-Fold 時，訓練集的資料都不可以比測試集的資料新。
3. k 值越大，得到誤差越小、但是需要運算的時間變長。因此，條件許可下，建議至少使用 k = 10. 或是使用  leave-one-out cross-validation (k = number of data instances)
